```python
# for all
for i in VarList:
    print("==================================================")
    print("Two-way Mixed ANOVA with "+i)
    print("==================================================")
    print("Mixed ANOVA")
    print(pg.mixed_anova(data=df, dv=i, between='Group', within='Treatment', subject='Participant').to_markdown(tablefmt="grid"))
    print("Pairwise ttests, Between first")
    print(pg.pairwise_tests(data=df, dv=i, between='Group', within='Treatment', subject='Participant').to_markdown(tablefmt="grid"))
    print("Pairwise ttests, Within first")
    print(pg.pairwise_tests(data=df, dv=i, between='Group', within='Treatment', subject='Participant', within_first=False).to_markdown(tablefmt="grid"))
    print("Means & SD")
    print(df.groupby(['Treatment','Group'])[i].describe().to_markdown(tablefmt="grid"))
    print(df.groupby(['Group'])[i].describe().to_markdown(tablefmt="grid"))
    print(df.groupby(['Treatment'])[i].describe().to_markdown(tablefmt="grid"))
    print("      ")
    print("      ")
```
