# Tests

## Depression as Dependent Variable

```r
Depression1 <- lm(
    Depression ~ 
    SINGLE+
    GENDER+AGE+GRADE+EDU+INCOME+HAPPYWITHPARENTS
    , data = df)
summary(Depression1)
```

```r
Depression2 <- lm(
    Depression ~ 
    COOP+
    GENDER+AGE+GRADE+EDU+INCOME+HAPPYWITHPARENTS
    , data = df)
summary(Depression2)
```

```r
Depression3 <- lm(
    Depression ~ 
    ONLINE+
    GENDER+AGE+GRADE+EDU+INCOME+HAPPYWITHPARENTS
    , data = df)
summary(Depression3)
```
