```python
df=pd.read_csv('plspm_csv.csv')
```

```python
print(df.columns.values)
```

```python
#subdataframes for Demographics, TextType, and Eyewear
dfD=df[['Gender', 'TextType', 'GC', 'IMG', 'Role', 'Nativeness', 'Group', 'Age','Eyewear']]
#create a list for T_ and S_ columns
T_list=df.columns[pd.Series(df.columns).str.startswith('T_')]
S_list=df.columns[pd.Series(df.columns).str.startswith('S_')]
#subdataframes for T_ and S_ columns
dfT=df[T_list]
dfS=df[S_list]
#merge the subdataframes with Demographics
dfTD=pd.merge(dfD,dfT, left_index=True, right_index=True)
dfSD=pd.merge(dfD,dfS, left_index=True, right_index=True)
#Overwrite the original dataframe with the new subdataframes
dfT=dfTD
dfS=dfSD
```

```python
#Running EFA
fa=FactorAnalyzer(n_factors=6, rotation="varimax")#setting terms
fa.fit(df[T_list])#applying terms for df with target variables
ev, v = fa.get_eigenvalues()#eigenvalues
EFAloaging=pd.DataFrame(fa.loadings_)#creating EFA loading df
EFAloaging['Variables'] = df[T_list].columns#setting index
EFAloaging.to_csv('EFAloaging.csv')#saving EFA loading df
print('eigenvalues')
pd.DataFrame(ev)
# print('EFA loading')
# EFAloaging
EFAloaging.to_csv('EFAloaging_T.csv')#saving EFA loading df
```

```python
#Running EFA
fa=FactorAnalyzer(n_factors=6, rotation="varimax")#setting terms
fa.fit(df[S_list])#applying terms for df with target variables
ev, v = fa.get_eigenvalues()#eigenvalues
EFAloaging=pd.DataFrame(fa.loadings_)#creating EFA loading df
EFAloaging['Variables'] = df[S_list].columns#setting index
EFAloaging.to_csv('EFAloaging.csv')#saving EFA loading df
print('eigenvalues')
pd.DataFrame(ev)
# print('EFA loading')
# EFAloaging
EFAloaging.to_csv('EFAloaging_S.csv')#saving EFA loading df
```

```python
#Creating subdataframes for each factor
satisfaction=['T_Satisfaction1','T_Intention2','T_Intention4','T_Confidence1','T_Confidence2','T_Confidence5','T_Satisfaction6','T_Satisfaction4','T_Satisfaction2']
presence=['T_Presence1','T_Presence9','T_Presence2','T_Presence11','T_Presence10','T_Presence3','T_Presence6','T_Presence5']
flow=['T_Flow1','T_Flow2','T_Flow3','T_Flow4','T_Flow6']
halo=['T_Halo1','T_Halo2','T_Halo3','T_Halo4']
memory=['T_memory', 'T_Recall', 'T_RecallRecog1', 'T_Recog1']
confidence=['T_Confidence3','T_Confidence4','T_Confidence6','T_Confidence7','T_Confidence8']
```

```python
#Cronbach's alpha for each factor
factorlist = [satisfaction, presence, flow, halo, memory, confidence]
for factor in factorlist:
    Cronbach = dfT[factor]
    print(factor)
    print(pg.cronbach_alpha(data=Cronbach))
```

```python
#Creating Composite Variables
dfT['satisfaction'] = dfT[satisfaction].mean(axis=1)
dfT['presence'] = dfT[presence].mean(axis=1)
dfT['flow'] = dfT[flow].mean(axis=1)
dfT['halo'] = dfT[halo].mean(axis=1)
dfT['memory'] = dfT[memory].mean(axis=1)
dfT['confidence'] = dfT[confidence].mean(axis=1)
```

```python
CompList=['satisfaction','presence','flow','halo','memory','confidence']
# Define the IV and DV variables
IV = 'Group'
DV = CompList
DF = dfT
# Run the ANOVA
for i in DV:
    anova_result = pg.anova(data=DF, dv=i, between=IV)
    globals()[f"ANOVA_{i}"] = anova_result
    globals()[f"ANOVA_{i}"].insert(0, 'Variable', i)
    pairwise_comparisons = pg.pairwise_tests(data=DF, dv=i, between=IV, padjust='bonf')
    globals()[f"Pairwise_{i}"] = pairwise_comparisons
    globals()[f"Pairwise_{i}"].insert(0, 'Variable', i)
    mean_table = DF.groupby(IV)[i].mean().reset_index()
    std_table = DF.groupby(IV)[i].std().reset_index()
    mean_std_table = pd.merge(mean_table, std_table, on=IV)
    mean_std_table.columns = [i, 'Mean', 'Standard Deviation']
    mean_std_table = mean_std_table.rename(columns={i: 'Value'})
    globals()[f"mean_std_table_{i}"] = mean_std_table
    globals()[f"mean_std_table_{i}"].insert(0, 'Variable', i)
merged_anova = pd.concat([globals()[f"ANOVA_{i}"] for i in DV], axis=0)
merged_anova.reset_index(drop=True, inplace=True)
merged_pairwise = pd.concat([globals()[f"Pairwise_{i}"] for i in DV], axis=0)
merged_pairwise.reset_index(drop=True, inplace=True)
merged_mean_std_table = pd.concat([globals()[f"mean_std_table_{i}"] for i in DV], axis=0)
merged_mean_std_table.reset_index(drop=True, inplace=True)
merged_anova
merged_pairwise
merged_mean_std_table
# merged_anova.to_csv(IV+'/merged_anova.csv', index=False)
# merged_pairwise.to_csv(IV+'/merged_pairwise.csv', index=False)
# merged_mean_std_table.to_csv(IV+'/merged_mean_std.csv', index=False)
```
